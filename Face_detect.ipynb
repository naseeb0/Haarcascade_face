```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
naseeb_face = cv2.imread('nass.png',0)
```

```python
plt.imshow(naseeb_face, cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_default.xml')
```

```python
def face_detect(img):
    face_img = img.copy()
    
    face_rect = face_cascade.detectMultiScale(face_img)
    for (x,y,w,h) in face_rect:
        cv2.reactangle(face_img, (x,y),(x+w,y+h),(255,255,255),10)
    return face_img
```

```python
final_img = face_detect(naseeb_face)
```

```python
plt.imshow(final_img,cmap='gray')
```
